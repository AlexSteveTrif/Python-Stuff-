```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Basic Graph


```python
x = [1,2,3]
y= [2,4,6]

# resize graph
plt.figure(figsize=(5,3), dpi = 500)



# Line number 1

plt.plot(x, y, label='2x', color = 'cyan', linewidth = 2, marker= '*', markeredgecolor= 'green', linestyle = '--')

# Line Number 2

x2 = np.arange(1,4.5,0.5)

plt.plot(x2, x2**2, 'magenta', label = 'x^2')

plt.plot(x2[4:], x2[4:])

plt.title('Our Fist Graph', fontdict={'fontname': 'Times New Roman', 'fontsize': 30})
plt.xlabel('X AXXXIS')
plt.ylabel('Y AXISSS')

plt.xticks([1,2,3])
plt.yticks([2,4,6])

plt.legend()

plt.show()

plt.savefig('mygiraffe.png', dpi=500)
```

# Bar Charts

```python
labels = ['A', 'B', 'C']
values = [ 1, 4, 2]

bars = plt.bar(labels,values)

bars[0].set_hatch('/')
bars[1].set_hatch('O')
bars[2].set_hatch('*')

plt.figure(figsize=(6,4))

plt.show()
```
